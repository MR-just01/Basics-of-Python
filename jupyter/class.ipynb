__init__ function is the constructor .. all classes have this func which is always excecuted when the class is being initiated 

now here comes the self parameter is a reference to the current instance(object) of the class, and is used to access variable that belongs to the class 

```python
class student:
    def __init__(self , name , age , course):
        print("init method is called i.e constructor is called")
        self.fullname = name 
        self.stdage = age
        self.stdcourse = course

s3 = student("abcs " , 24,"python")
print(s3.fullname)
print(s3.stdage )
print(s3.stdcourse)

s4 = student("no " , 22,"python")
print(s4.fullname)
print(s4.stdage )
print(s4.stdcourse)

```

STATIC METHOD
methods that don't use the self parameter (work at the class level )
syntax : 
@staticmethod 


###Decorator allow us to wrap another function in order to extends the behaviour of the wrapped function, without permanently modifying it 

```python
class kids:
    def __init__(self , name , sub1, sub2, sub3 ):
        self.name = name
        self.sub1 = sub1
        self.sub2 = sub2
        self.sub3 = sub3


    @staticmethod
    def greet():
        print("annyeonghaseyo")

    


    def average(self):
        average = (self.sub1 +self.sub2+self.sub3)/3
        print("name of the student is : " , self.name )
        print("average marks of 3 subjects is : " , average )


s1 = kids("Maaya " , 12 , 13, 15)
s1.average()
s1.greet()



```

ABSTRACTION : IT IS THE PILLAR OF THE OOPS THAT HELP IN THE HIDDING THE NON-TRIVIAL IMPLEMENTATION FROM THE CLIENT



```python
class car :
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False

    def start(self):
        self.clutch = True 
        self.acc = True 
        print("car  started...")

car1 = car()
car1.start()
```

ENCAPSULATION  = IT IS THE WRAPPING UP THE DATA OR ATTIRIBUTES AND METHODS IN THE SINGLE UNIT 


```python
##PRACTICE 
#CREATE ACCOUNT CLASS WITH 2 ATTRIBUTES - BALANECE AND ACCOUNR NUMNER 
#CREATE THE METHOD TO DEBIT , CREDIT AND PRINTING THE BALLANCE

class account:
    def __init__(self , acc_num , balance):
        self.acc_num = acc_num
        self.balance = balance 
    
    def debit(self ,amount):
        if self.balance >= amount:
            self.balance = self.balance - amount
            print("amount debited !!!")
        else:
            print("insufficient balance")
    
    def credit(self , amount):
        self.balance = self.balance + amount
        print("amount credited!!")

    def print_balance(self):
        print("your account number is : " , self.acc_num)
        print("your current balance is : " , self.balance)

 
acc1 = account(123456 , 10000)
acc1.print_balance()
acc1.debit(2000)
acc1.print_balance()
acc1.credit(5000)
acc1.print_balance()


```
